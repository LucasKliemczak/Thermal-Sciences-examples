# Flow Over Cylinder
$$
Vr =  \frac{‚àÇùúô}{‚àÇr} =U(1-\frac{R^2}{r^2})cos\theta     
$$
*and*
$$
V_{\theta} =  \frac{1}{r}\frac{‚àÇùúô}{‚àÇùúÉ} = -U(1-\frac{R^2}{r^2})sin\theta     
$$
##  .
$$
  \left[ \begin{array}{c} V_x \\ V_y \end{array} \right] =
  \left[ \begin{array}{cc} cosùúΩ & -sinùúΩ \\ sinùúΩ & cosùúΩ \end{array}\right]
  \left[ \begin{array}{c} V_r \\ V_ùúÉ \end{array} \right]
$$
** https://en.wikipedia.org/wiki/Potential_flow_around_a_circular_cylinder **

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,200) #gridsize
y = np.linspace(-5,5,200)
X,Y = np.meshgrid(x,y)

r = np.sqrt(x**2 + Y**2) # radial distance
theta = np.arctan(Y/X) # angle

U = 1 # flow velocity
R = 1.5 # Radius of the cylinder

Vr = U*(1 - (R**2/r**2))*np.cos(theta) # radial
Vt = -U*(1 + (R**2/r**2))*np.sin(theta) # tangential

Vx = Vr*np.cos(theta) - Vt*np.sin(theta)
Vy = Vr*np.sin(theta) + Vt*np.cos(theta)

plt.figure(figsize=(6,5),dpi=100)
circle1 = plt.Circle((0,0),R,color='grey')
plt.gca().add_patch(circle1)
Vx[r<=R]=np.nan
Vy[r<=R]=np.nan
# alpha = np.linspace(0,2*np.pi,360)
# xp = R*np.cos(alpha)
# yp = R.np.sin(alpha)
# plt.plot(xp,yp,clor='black')
plt.colorbar(plt.streamplot(X,Y,Vx,Vy,color=np.sqrt(Vx**2+Vy**2),density = [1.5,1.5], cmap = "jet").lines)
plt.show()
```
