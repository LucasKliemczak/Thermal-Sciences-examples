```python
import numpy as np
import matplotlib.pyplot as plt

n = 5

x = np.linspace(-n, n, 100)
y = np.linspace(-n, n, 100)

X,Y = np.meshgrid(x,y)

U = X**2 + Y - 1
V = X + Y**2 + 1

plt.figure(figsize=(5,5))
vel_streams = plt.streamplot(X,Y,U,V,color = U, density = [1,1], cmap = "jet")
plt.colorbar((vel_streams).lines)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 5
x = np.linspace(-n, n, 5)
y = np.linspace(-n, n, 5)

X, Y = np.meshgrid(x,y)

U = x**2 + Y -1 # U component of fluid vel
V = X + Y**2 +1 # V component of fluid vel

plt.figure(figsize=(6,6))
fluid_streams = plt.streamplot(X, Y, U, V, color=np.sqrt(U**2 + V**2), density=[1,1], cmap='jet')
plt.colorbar((fluid_streams).lines)
plt.show()

```
